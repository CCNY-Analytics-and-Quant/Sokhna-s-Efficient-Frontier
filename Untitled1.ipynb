```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
ticker = "SPY"
etf = yf.Ticker(ticker)
asset = etf.ticker
etf_data = etf.history(period="5y")
data = yf.download(asset, period="5y")
```

```python
momentum_factors = pd.DataFrame()
etf_returns = etf_data.pct_change()
momentum_factors = (etf_returns + 1).rolling(12).apply(np.prod, raw=True) - 1
```

```python
momentum_factors
```

```python
z_scores = (momentum_factors - momentum_factors.mean()) / momentum_factors.std()
```

```python
z_scores
```

```python
long_basket = z_scores.mean(axis=1).nlargest(15).index
```

```python
short_basket = z_scores.mean(axis=1).nsmallest(15).index.tolist()
```
