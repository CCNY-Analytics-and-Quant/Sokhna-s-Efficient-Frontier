```python
import yfinance as yf
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import datetime as dt
import math
```

```python
assets = ['AMD','AAPL','TSLA','BAC','HD','NKE','KO']
period = '2y'
Adjclose = yf.download(assets,period = period)['Adj Close']
```

```python
returns=Adjclose.pct_change()
table_1 = pd.DataFrame(index=assets)
table_1['Variance']=(Adjclose.pct_change()*100).var()
table_1['Mean']=returns.mean()
```

```python
import numpy as np
def port_ret(weights):
    return np.sum(returns.mean() * weights) * 252
```

```python
def port_vol(weights):
    return np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
```

```python
portfolio = len(assets)
prets = []
pvols = []
for p in range (25000):    
    weights = np.random.random(portfolio)                   
    weights /= np.sum(weights)                               
    prets.append(port_ret(weights)) 
    pvols.append(port_vol(weights))
prets = np.array(prets)
pvols = np.array(pvols)

```

```python
from pylab import mpl, plt
plt.figure(figsize=(10, 6))
plt.scatter(pvols, prets, c=prets / pvols,marker='o', cmap='coolwarm')
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio');
```

```python
import scipy.optimize as sco
def min_func_sharpe(weights): 
    return -port_ret(weights) / port_vol(weights)
```

```python
cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bnds = tuple((0, 1) for x in (assets)) 
```

```python
eweights = np.array(portfolio * [1. / portfolio,])
```

```python
min_func_sharpe(eweights)
```

```python
 opts = sco.minimize(min_func_sharpe, eweights,
 method='SLSQP', bounds=bnds,
 constraints=cons) 
 
```
